## ¿Qué es la ciencia de datos?
HayUn chiste que dice que un científico de datos es alguien que sabe más estadística que un informático y más informática que un estadístico. (No dije que fuera un buen chiste). De hecho, algunos científicos de datos son, a efectos prácticos, estadísticos, mientras que otros son prácticamente indistinguibles de los ingenieros de software. Algunos son expertos en aprendizaje automático, mientras que otros no podrían aprenderlo ni siquiera en el jardín de infancia. Algunos tienen doctorados con impresionantes historiales de publicaciones, mientras que otros nunca han leído un artículo académico (¡qué vergüenza!). En resumen, prácticamente sin importar cómo se defina la ciencia de datos, siempre habrá profesionales para quienes la definición es totalmente errónea.

Sin embargo, eso no nos impedirá intentarlo. Podríamos decir que un científico de datos es alguien que extrae información valiosa de datos desordenados. El mundo actual está lleno de personas que intentan convertir datos en conocimiento.

Por ejemplo, el sitio de citas OkCupid pide a sus miembros que respondan miles de preguntas para encontrar las parejas más adecuadas. Pero también analiza estos resultados para elaborar preguntas aparentemente inofensivas que puedes hacerle a alguien para averiguar la probabilidad de que se acueste contigo en la primera cita .

Facebook te pide que indiques tu ciudad natal y tu ubicación actual, supuestamente para que a tus amigos les resulte más fácil encontrarte y conectar contigo. En 2012, la campaña de Obama empleó a decenas de científicos de datos que, mediante la minería de datos y la experimentación, identificaron a los votantes que necesitaban mayor atención, seleccionaron las campañas y programas de recaudación de fondos más adecuados para cada donante y centraron los esfuerzos para movilizar el voto donde resultaran más efectivos. ## Encontrar conectores clave
Es tu primer día de trabajo en DataSciencester, y el vicepresidente de redes está lleno de preguntas sobre tus usuarios. Hasta ahora no había tenido a quién preguntar, así que está muy contento de tenerte en el equipo.

En concreto, quiere que identifiques quiénes son los "conectores clave" entre los científicos de datos. Para ello, te proporciona un volcado de toda la red de DataSciencester. (En la práctica, no es habitual que te den los datos que necesitas. El capítulo 9 está dedicado a la obtención de datos).

¿Qué aspecto tiene este volcado de datos? Consiste en una lista de usuarios, cada uno representado por un `dict` (diccionario en Python) que contiene el `id`(que es un número) de ese usuario y `name`(que, en una de las grandes coincidencias cósmicas, rima con el del usuario id):

```python
users = [
    { "id": 0, "name": "Hero" },
    { "id": 1, "name": "Dunn" },
    { "id": 2, "name": "Sue" },
    { "id": 3, "name": "Chi" },
    { "id": 4, "name": "Thor" },
    { "id": 5, "name": "Clive" },
    { "id": 6, "name": "Hicks" },
    { "id": 7, "name": "Devin" },
    { "id": 8, "name": "Kate" },
    { "id": 9, "name": "Klein" }
]
```

También te proporciona los datos de “amistad”, representados como una lista de pares de identificadores:

```python
friendship_pairs = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
                    (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

Por ejemplo, la tupla (0, 1)indica que el científico de datos con id0 (Hero) y el científico de datos con id1 (Dunn) son amigos. La red se ilustra en la Figura 1-1 .

 ![descripcion](./images/dsf2_0101.png)

Representar las amistades como una lista de pares no es la forma más sencilla de trabajar con ellas. Para encontrar todas las amistades del usuario 1, hay que recorrer cada par buscando aquellos que contengan el número 1. Si hubiera muchos pares, esto llevaría mucho tiempo.

En cambio, vamos a crear una tabla dictdonde las claves sean usuarios idy los valores sean listas de amigos id. (Buscar información en una tabla dictes muy rápido).

Todavía tendremos que examinar cada par para crearlo dict, pero solo tenemos que hacerlo una vez, y después obtendremos búsquedas económicas:

```python
#Initialize the dict with an empty list for each user id:
friendships = {user["id"]: [] for user in users}
friendships
```

```python
# And loop over the friendship pairs to populate it:
for i, j in friendship_pairs:
    friendships[i].append(j) # Add j as a friend of user i
    friendships[j].append(i) # Add i as a friend of user j

friendships
```

Ahora que tenemos las amistades en un `dict`, podemos hacer preguntas fácilmente a nuestro gráfico, como "¿Cuál es el número promedio de conexiones?"

Primero hallamos el número total de conexiones, sumando las longitudes de todas las `friends` listas:

```python
def number_of_friends(user):
    """How many friends does _user_ have?"""
    user_id = user["id"]
    friends_ids = friendships[user_id]
    return len(friends_ids)

total_connections = sum(number_of_friends(user) for user in users)
total_connections
```

Y luego simplemente dividimos por el número de usuarios:

```python
num_users = len(users)                            # length of the users list
avg_connections = total_connections / num_users   # 24 / 10 == 2.4
avg_connections
```

También es fácil encontrar a las personas con más contactos: son aquellas que tienen mayor número de amigos.

Como no hay muchos usuarios, podemos simplemente ordenarlos de "más amigos" a "menos amigos":

```python
# Create a list(user_id, number_of_friends)
num_friends_by_id = [(user["id"], number_of_friends(user)) for user in users]

num_friends_by_id.sort(
    key=lambda id_and_friends: id_and_friends[1],
    reverse=True
)
```

```python
num_friends_by_id
```

Podemos entender lo que hemos hecho como una forma de identificar a las personas que son, de alguna manera, fundamentales para la red. De hecho, lo que acabamos de calcular es la métrica de centralidad de grado de la red ( Figura 1-2 ).

![descripcion](./images/dsf2_0102.png)

Esto tiene la ventaja de ser bastante fácil de calcular, pero no siempre da los resultados que uno desearía o esperaría. Por ejemplo, en la red DataSciencester, Thor(`id` 4) solo tiene dos conexiones, mientras que Dunn(`id`1) tiene tres. Sin embargo, al observar la red, intuitivamente parece que Thor debería ser más central. En el Capítulo 22 , investigaremos las redes con más detalle y analizaremos nociones más complejas de centralidad que pueden o no ajustarse mejor a nuestra intuición.

## Científicos de datos que quizás conozcas

Mientras sigues rellenando los formularios de contratación, la vicepresidenta de Relaciones Interpersonales se acerca a tu escritorio. Quiere fomentar una mayor interacción entre los miembros y te pide que diseñes un sistema de sugerencias llamado "Científicos de Datos que Quizás Conozcas".

Tu primer instinto es sugerir que los usuarios podrían conocer a los amigos de sus amigos. Así que escribes un código para iterar sobre sus amigos y recopilar los amigos de los amigos:

```python
def foaf_ids_bad(user):
    """foaf is short for `friend of a friend` """
    return [foaf_id
            for friend_id in friendships[user["id"]]
            for foaf_id in friendships[friend_id]
           ]
```

Cuando llamamos a esto en `users[0]` (Hero), produce:

```python
[0, 2, 3, 0, 1, 3]
```

Incluye al usuario 0 dos veces, ya que Hero es amigo de ambos. Incluye a los usuarios 1 y 2, aunque ambos ya son amigos de Hero. Y también incluye al usuario 3 dos veces, puesto que se puede contactar con Chi a través de dos amigos diferentes.
